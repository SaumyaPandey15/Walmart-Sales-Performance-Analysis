```python
import pandas as pd
import sqlite3
```

```python
df = pd.read_csv("cleaned_sales_data.csv")
```

```python
conn = sqlite3.connect("walmart_sales.db")
```

```python
df.to_sql(
    name="sales",
    con=conn,
    if_exists="replace",
    index=False
)
```

```python
pd.read_sql("""
SELECT COUNT(*) AS total_records
FROM sales;
""", conn)
```

```python
pd.read_sql("""
SELECT ROUND(SUM(Weekly_Sales), 2) AS total_revenue
FROM sales;
""", conn)
```

```python
pd.read_sql("""
SELECT ROUND(AVG(Weekly_Sales), 2) AS average_weekly_sales
FROM sales;
""", conn)
```

```python
pd.read_sql("""
SELECT
    Store,
    ROUND(SUM(Weekly_Sales),2) AS revenue
FROM sales
GROUP BY Store
ORDER BY revenue DESC
LIMIT 10;
""", conn)
```

```python
pd.read_sql("""
SELECT
    Holiday_Flag,
    ROUND(SUM(Weekly_Sales),2) AS revenue,
    COUNT(*) AS total_weeks
FROM sales
GROUP BY Holiday_Flag;
""", conn)
```

```python
pd.read_sql("""
SELECT
    Store,
    ROUND(AVG(Weekly_Sales),2) AS avg_sales
FROM sales
GROUP BY Store
ORDER BY avg_sales DESC;
""", conn)
```

```python
pd.read_sql("""
SELECT
    Store,
    ROUND(SUM(Weekly_Sales),2) AS revenue
FROM sales
GROUP BY Store
HAVING SUM(Weekly_Sales) >
(
    SELECT AVG(Weekly_Sales) * COUNT(DISTINCT Store)
    FROM sales
)
ORDER BY revenue DESC;
""", conn)
```
